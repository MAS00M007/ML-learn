```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds =  pd.read_csv("loan.csv")
ds.head(3)
```

```python
ds.isnull().sum()
```

```python
ds.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
ms.fit(ds[["CoapplicantIncome"]])
```

```python
ds["CoapplicantIncome_minMAX"] = ms.transform(ds[["CoapplicantIncome"]])
```

```python
ds
```

```python
plt.figure(figsize=(13,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(ds["CoapplicantIncome"])



plt.subplot(1,2,2)
plt.title("After")
sns.distplot(ds["CoapplicantIncome_minMAX"])
plt.show()
```

```python
import pandas as pd
```

```python
ds = pd.read_csv("Social_Network_Ads.csv")
ds.head(20)
```

```python
ds["Purchased"].value_counts()
```

```python
x = ds.iloc[:,:-1]
y = ds["Purchased"]
```

```python
!pip install imblearn
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
ru = RandomUnderSampler()
ru_x,ru_y = ru.fit_resample(x,y)
```

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(ru_x,ru_y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ro = RandomOverSampler()
ro_x,ro_y = ro.fit_resample(x,y)
```

```python
ro_y.value_counts()
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(ro_x,ro_y,test_size=0.20,random_state=42)
```

```python
lr1 = LogisticRegression()
lr1.fit(x_train1,y_train1)
```

```python
lr.score(x_test1,y_test1)*100
```

```python
lr.predict([[48,29000]])
```

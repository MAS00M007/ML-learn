```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("Social_Network_Ads.csv")
ds.drop(columns=["EstimatedSalary"],inplace=True)
ds.head(3)
```

```python
ds.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="Age",y="Purchased",data=ds)
plt.show()
```

```python
x = ds.iloc[:,:-1]
y = ds["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.20,random_state=42)  
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[32]])
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x="Age",y="Purchased",data=ds)
sns.lineplot(x = "Age",y=lr.predict(x),data=ds,color='red')
plt.show()
```

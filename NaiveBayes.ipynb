```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
ds = pd.read_csv("placement.csv")
ds.head(3)
```

```python
ds.isnull().sum()
```

```python
sns.scatterplot(x="cgpa",y="iq",data=ds,hue="placement")
plt.show()
```

```python
# to check if it is noraml destribution or not i.e is it forming a bell shape
sns.kdeplot(data=ds["cgpa"])
plt.show()
```

```python
x=ds.iloc[:,:-1]
y=ds["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gNB = GaussianNB()
gNB.fit(x_train,y_train)
```

```python
gNB.score(x_test,y_test)*100
```

```python
mNB = MultinomialNB()
mNB.fit(x_train,y_train)
```

```python
mNB.score(x_test,y_test)*100
```

```python
bNB = BernoulliNB()
bNB.fit(x_train,y_train)
```

```python
bNB.score(x_test,y_test)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gNB)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mNB)
plt.show()
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bNB)
plt.show()
```

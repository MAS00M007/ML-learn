```python
q1 = ds["CoapplicantIncome"].quantile(0.25)
q3 = ds["CoapplicantIncome"].quantile(0.75)
```

```python
IQR= q3-q1
IQR
```

```python
min_range = q1-(1.5*IQR)
max_range = q3+(1.5*IQR)
min_range,max_range
```

```python
ds
```

```python
newds = ds[ds["CoapplicantIncome"]<=max_range]
```

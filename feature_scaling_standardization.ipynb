```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("loan.csv")
ds.head(3)
```

```python
ds["ApplicantIncome"].isnull().sum()
```

```python
sns.distplot(ds["ApplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(ds[["ApplicantIncome"]])
```

```python
ds["ApplicantIncome_ss"]= pd.DataFrame(ss.transform(ds[["ApplicantIncome"]]),columns=["x"])
```

```python
ds.describe()
```

```python
plt.figure(figsize=(12,5))
plt.title("Before")
plt.subplot(1,2,1)
sns.distplot(ds["ApplicantIncome"])



plt.subplot(1,2,2)
plt.title("After")
sns.distplot(ds["ApplicantIncome_ss"])
plt.show()
```

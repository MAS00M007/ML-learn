```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("polynomial_regression.csv")
ds.head(3)
```

```python
ds.head(3)
```

```python
x = ds.iloc[:,:-1]
y = ds["Ice Cream Sales (units)"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(criterion ='squared_error', max_depth =  14, splitter = 'random')
dtr.fit(x_train,y_train)
```

```python
dtr.score(x_test,y_test)*100,dtr.score(x_train,y_train)*100
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
df = {"criterion":["squared_error", "friedman_mse", "absolute_error","poisson"],
    "splitter":["best","random"],
    "max_depth":[i for i in range(2,20)]}
```

```python
gd = GridSearchCV(DecisionTreeRegressor(),param_grid=df,cv=10)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_
```

```python
gd.cv_results_
```

```python
rcv = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
rcv.fit(x_train,y_train)
```

```python
rcv.best_score_
```

```python
rcv.best_params_
```

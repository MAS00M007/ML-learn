```python
import pandas as pd
```

```python
ds = pd.read_csv("loan.csv")
ds.head(3)
```

```python
ds.isnull().sum()
```

```python
ds["Gender"]=ds["Gender"].fillna(ds["Gender"].mode()[0])
```

```python
ds.isnull().sum()
```

```python
ds["Married"]=ds["Married"].fillna(ds["Married"].mode()[0])
```

```python
ds.isnull().sum()
```

```python
en_ds = ds[["Gender","Married"]]
en_ds
```

```python
pd.get_dummies(en_ds)
```

```python
pd.get_dummies(en_ds).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first")
arr = ohe.fit_transform(en_ds).toarray()
```

```python
pd.DataFrame(arr,columns=["Gender_Female","Married_No"])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = {
    'CGPA': [9.1, 8.5, 6.8, 7.2, 8.9, 7.5, 9.4, 6.5, 8.1, 7.0, 8.7, 6.9, 7.9, 9.2, 6.4],
    'Package': [12.0, 9.5, 4.2, 5.1, 11.0, 6.0, 13.5, 3.8, 8.0, 4.8, 10.0, 4.0, 7.5, 12.5, 3.5]
}
df = pd.DataFrame(data)
```

```python
df.head(4)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x="CGPA",y="Package",data=df)
plt.show()
```

```python
df.info()
```

```python
x = df[["CGPA"]]
y = df[["Package"]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[9]])
```

```python
y_pred = lr.predict(x)
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x="CGPA",y="Package",data=df)
plt.plot(df["CGPA"],y_pred,c="red")
plt.legend(["org Data","prdict line"])
plt.savefig("fig.jpg")
plt.show()
```
